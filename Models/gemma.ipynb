```python
# Import the necessary libraries
from transformers import pipeline, AutoTokenizer, AutoModelForCausalLM

# Define a pipeline for text generation
pipe = pipeline("text-generation", model="google/codegemma-2b")

# Define a prompt for code generation
prompt = "Write a Python function that computes the factorial of a number:"

# Generate text
output = pipe(prompt, max_length=100, num_return_sequences=1)

# Clean the generated text by filtering out unwanted tokens
cleaned_output = output[0]['generated_text'].replace('<|fim_prefix|>', '').replace('<|fim_suffix|>', '').replace('<|fim_middle|>', '')

# Print the cleaned output
print(cleaned_output)

# Alternatively, loading the model and tokenizer directly
# Load the tokenizer
tokenizer = AutoTokenizer.from_pretrained("google/codegemma-2b")

# Load the model directly
model = AutoModelForCausalLM.from_pretrained("google/codegemma-2b")

# Encode the input prompt
input_ids = tokenizer.encode(prompt, return_tensors="pt")

# Generate output from the model
output = model.generate(input_ids, max_length=100, num_return_sequences=1)

# Decode and clean the output text
generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
cleaned_generated_text = generated_text.replace('<|fim_prefix|>', '').replace('<|fim_suffix|>', '').replace('<|fim_middle|>', '')

# Print the cleaned generated output
print(cleaned_generated_text)
```
